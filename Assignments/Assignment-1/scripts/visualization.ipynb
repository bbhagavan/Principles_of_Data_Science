```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('../data/StudentsPerformance.csv')

# 1. Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['math score'], bins=10, kde=True)
plt.title('Distribution of Math Scores')
plt.savefig('../output/visualizations/histogram.png')

# 2. Bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='math score', data=df)
plt.title('Gender vs Math Scores')
plt.savefig('../output/visualizations/bar_plot.png')

# 3. Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='reading score', y='writing score', data=df)
plt.title('Reading vs Writing Scores')
plt.savefig('../output/visualizations/scatter_plot.png')

# 4. Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.title('Parental Level of Education vs Math Scores')
plt.xticks(rotation=45)
plt.savefig('../output/visualizations/box_plot.png')

# 5. Heatmap
corr_matrix = df[['math score', 'reading score', 'writing score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('../output/visualizations/heatmap.png')
```
