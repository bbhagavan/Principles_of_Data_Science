```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../data/cleaned_data.csv')

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Grip strength', hue='Frailty', data=df)
plt.title('Grip Strength vs. Age')
plt.savefig('../output/eda/grip_strength_vs_age.png')

correlation_matrix = df[['Height', 'Weight', 'Age', 'Grip strength']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('../output/eda/correlation_matrix.png')
```
